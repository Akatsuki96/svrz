```python
import numpy as np
import matplotlib.pyplot as plt
```

# Linear Regression

```python
out_path = "./results/linear_regression"
```

```python
def read_result(fname):
    mu_val, mu_times, std_val, std_times = [], [], [], []
    with open(f"{fname}", 'r') as f:
        lines = f.readlines()
    for line in lines:
        mu_fx, mu_tm, std_fx, std_tm = line.split(",")
        mu_val.append(float(mu_fx))
        mu_times.append(float(mu_tm))
        std_val.append(float(std_fx))
        std_times.append(float(std_tm))
    return np.array(mu_val), np.array(mu_times), np.array(std_val), np.array(std_times)
```

```python
mu_sszd_values, mu_sszd_times, std_sszd_values, std_sszd_times = read_result(f"{out_path}/sszd.log")
mu_svrz_values, mu_svrz_times, std_svrz_values, std_svrz_times = read_result(f"{out_path}/svrz.log")
```

```python
fig, ax = plt.subplots()

ax.plot(range(len(mu_sszd_values)), mu_sszd_values, '-', label='SSZD')
ax.fill_between(range(len(mu_sszd_values)), abs(mu_sszd_values - std_sszd_values) , mu_sszd_values + std_sszd_values, alpha=0.6 )
ax.plot(range(len(mu_svrz_values)), mu_svrz_values, '-', label='SVRZ')
ax.fill_between(range(len(mu_svrz_values)), abs(mu_svrz_values - std_svrz_values) , mu_svrz_values + std_svrz_values, alpha=0.6 )

ax.set_yscale("log")
ax.legend()

ax.set_title("Linear Regression")
ax.set_ylabel("$F(x_t) - F(x^*)$")
ax.set_xlabel("Number of Stochastic Function Evaluations")

fig.savefig("./linear_regression_values.pdf", bbox_inches='tight')
```

```python
fig, ax = plt.subplots()

ax.plot(mu_sszd_times, mu_sszd_values, '-', label='SSZD')
ax.fill_between(mu_sszd_times, abs(mu_sszd_values - std_sszd_values) , mu_sszd_values + std_sszd_values, alpha=0.6 )
ax.plot(mu_svrz_times, mu_svrz_values, '-', label='SVRZ')
ax.fill_between(mu_svrz_times, abs(mu_svrz_values - std_svrz_values) , mu_svrz_values + std_svrz_values, alpha=0.6 )

ax.set_yscale("log")
ax.legend()

ax.set_title("Linear Regression")
ax.set_ylabel("$F(x_t) - F(x^*)$")
ax.set_xlabel("Cumulative time (s)")

ax.set_xscale("log")

fig.savefig("./linear_regression_times.pdf", bbox_inches='tight')
```

```python
fig, ax = plt.subplots()

ax.plot(range(len(mu_sszd_times)), mu_sszd_times, '-', label='SSZD')
ax.fill_between(range(len(mu_sszd_times)), abs(mu_sszd_times - std_sszd_times) , mu_sszd_times + std_sszd_times, alpha=0.6 )
ax.plot(range(len(mu_svrz_times)), mu_svrz_times, '-', label='SVRZ')
ax.fill_between(range(len(mu_svrz_times)), abs(mu_svrz_times - std_svrz_times) , mu_svrz_times + std_svrz_times, alpha=0.6 )

ax.set_yscale("log")
ax.legend()

ax.set_title("Linear Regression")
ax.set_ylabel("Cumulative Time $(s)$")
ax.set_xlabel("Number of Stochastic Function Evaluations")

fig.savefig("./linear_regression_ctime.pdf", bbox_inches='tight')
```
