```python
import numpy as np
import matplotlib.pyplot as plt
```

# Linear Regression

```python
out_path = "./results/linear_regression"
```

```python
def read_result(fname):
    values, times = [], []
    with open(f"{fname}", 'r') as f:
        lines = f.readlines()
    for line in lines:
        value, tm = line.split(",")
        values.append(float(value))
        times.append(float(tm))
    return values, times
```

```python
sszd_values, sszd_times = read_result(f"{out_path}/sszd.log")
svrz_values, svrz_times = read_result(f"{out_path}/svrz.log")
```

```python
fig, ax = plt.subplots()

ax.plot(range(len(sszd_values)), sszd_values, '-', label='SSZD')
ax.plot(range(len(svrz_values)), svrz_values, '-', label='SVRZ')

ax.set_yscale("log")
ax.legend()

ax.set_title("Linear Regression")
ax.set_ylabel("$F(x_t) - F(x^*)$")
ax.set_xlabel("Number of Stochastic Function Evaluations")

fig.savefig("./linear_regression_values.pdf", bbox_inches='tight')
```

```python
fig, ax = plt.subplots()

ax.plot(sszd_times, sszd_values, '-', label='SSZD')
ax.plot(svrz_times, svrz_values, '-', label='SVRZ')

ax.set_yscale("log")
ax.legend()

ax.set_title("Linear Regression")
ax.set_ylabel("$F(x_t) - F(x^*)$")
ax.set_xlabel("Cumulative time (s)")

ax.set_xscale("log")

fig.savefig("./linear_regression_times.pdf", bbox_inches='tight')
```
